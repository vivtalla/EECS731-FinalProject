```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 500)
votesDF = pd.read_csv('../data/votes.csv', delimiter=",")
```

# FeatureEngineering

From the dataset there are two columns that seem useless (X and Unnamed: 0). I'll go ahead and remove them from our dataframe

```python
votesDF = votesDF.drop('X', axis=1)
```

```python
votesDF.head(5)
```

```python
votesDF = votesDF.drop('Unnamed: 0', axis=1)
```

```python
votesDF.head(5)
```

In this notebook I'm going use various features to predict the possible outcome of the election. In order to do so, I'll need to create a predictor variable, following the same steps as our previous notebook in this project. 

```python
for i,row in enumerate(votesDF.iterrows()):
    if row[1].votes_gop_2016>row[1].votes_dem_2016:
        votesDF.loc[i,'16Winner'] = 'Trump'
    else:
        votesDF.loc[i,'16Winner'] = 'Clinton'  
    if row[1].votes_gop_2012>row[1].votes_dem_2012:
        votesDF.loc[i,'12Winner'] = 'Romney'
    else:
        votesDF.loc[i,'12Winner'] = 'Obama' 
        
votesDF.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(votesDF['16Winner'])
votesDF['16Winner'] = le.transform(votesDF['16Winner'])
le.fit(votesDF['12Winner'])
votesDF['12Winner'] = le.transform(votesDF['12Winner'])

votesDF.head(5)
```

```python
votesDF.dtypes
votesDF["12Winner"] = pd.to_numeric(votesDF["12Winner"])
votesDF["16Winner"] = pd.to_numeric(votesDF["16Winner"])
votesDF.head(5)
```

To get a better idea which features will be best for a predictive model, we can do feature selection using an Extra Tree Classifier to identify which attributes contribute the most to predicting the 2016 election outcome.

# Feature Selection

```python
#Feature Importance with Extra Tree Classifier
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier

#Create Feature/Label Dataframes
X = votesDF[['population2014', 'population2010', 'AGE135214', 'AGE295214', 'age65plus', 'White', 'Black', 'Hispanic', 'NonEnglish', 'Edu_highschool', 'Edu_batchelors', '12Winner']]
Y = votesDF[['16Winner']]
X = np.array(X)
Y = np.array(Y) 

#Feature Extraction
model = ExtraTreesClassifier()
model.fit(X, Y)
print("Model Feature Importance:", model.feature_importances_)
type(model.feature_importances_)

import seaborn as sns

plt.figure(figsize=(18,10))
plt.title('Feature Importance with Extra Tree Classifier')
plt.xlabel('Features')
plt.ylabel('Score')
ETCFeatures = ['population2014', 'population2010', 'AGE135214', 'AGE295214', 'age65plus', 'White', 'Black', 'Hispanic', 'NonEnglish', 'Edu_highschool', 'Edu_batchelors', '12Winner']
sns.barplot(ETCFeatures, model.feature_importances_, palette='Blues_d')
```

We are given an importance score for each attribute. The larger score the more important the attribute. Based on the Extra Tree Classifier, the rankings for high scoring features as well as there score are as follows:

1: 2012 Election (.472)
2: Batchelors Education (.100)
3: % County Population White (.076)
4: % County Population Black (.074)

We see the 2012 Election results are the best feature for predicting the 2016 election outcome by a large margin. In another notebook, we will look at county turnover where a county who voted Democrat switches to Republican and visa-versa, but for now we will be sure to use that feature in our prediction models.

We can also model feature selection using Recursive Feature Elimination to identify which attributes contribute the most to predicting the 2016 election outcome.

```python
#Feature Extraction with RFE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

#Feature Extraction using RFE with the logistic Regression Algorithm
model = LogisticRegression()
rfe = RFE(model, 1)
rfe = rfe.fit(X, Y)


print("RFE n_Features:", rfe.n_features_)
print("RFE support:", rfe.support_)
print("RFE ranking: ", rfe.ranking_)
```

The RFE model provides us with a ranking for the features. The top 5 rankings are as follows:

1 - 2012 Election outcome
2 - Hispanic population
3 - White population
4 - Black population
5 - Batchelors Education

The top 5 features has the top 4 using Extra Tree Classifier, but in a different ranking. The scores in the extra tree classifier below the top 4 were all similair in value, so the changed in ranking likely has to do with the nuances of the individual algorithms.

# Decision Tree Classifier

Now that we have insight on which features are most important, we can use a Decision Tree Classifier to predict 2016 election outcome based off those features. I'm going to go ahead and predict using the different features individually, and then predict using all the important features.

```python
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree

labels = votes['16Winner']
features = votes[['Edu_batchelors']]
X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    labels, 
                                                    test_size=0.20, 
                                                    random_state=42)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(accuracy_score(y_test, clf.predict(X_test)))
```
