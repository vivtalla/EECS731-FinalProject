```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 500)
votesDF = pd.read_csv('../data/votes.csv', delimiter=",")
```

In this notebook I'm going use various features to predict the possible outcome of the election. In order to do so, I'll need to create a predictor variable, following the same steps as our previous notebook in this project. Refer to the python notebook using Random Forest classification to predict election outcomes/turnout, as it describes how we derive the predictor variables below.

```python
for i,row in enumerate(votesDF.iterrows()):
    if row[1].votes_gop_2016>row[1].votes_dem_2016:
        votesDF.loc[i,'16Winner'] = 'Trump'
    else:
        votesDF.loc[i,'16Winner'] = 'Clinton'  
    if row[1].votes_gop_2012>row[1].votes_dem_2012:
        votesDF.loc[i,'12Winner'] = 'Romney'
    else:
        votesDF.loc[i,'12Winner'] = 'Obama' 
        
votesDF.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(votesDF['16Winner'])
votesDF['16Winner'] = le.transform(votesDF['16Winner'])
le.fit(votesDF['12Winner'])
votesDF['12Winner'] = le.transform(votesDF['12Winner'])

votesDF.head(5)
```

```python
votesDF.dtypes
votesDF["12Winner"] = pd.to_numeric(votesDF["12Winner"])
votesDF["16Winner"] = pd.to_numeric(votesDF["16Winner"])
votesDF.head(5)
```

The dataframe has many features, but there are only so many I'm really interested in. I'm going to use my intuition to take features of interest, and then will plot a heatmap showing correlation of those features to better understand the dataset.

```python
votes = votesDF[['population2014', 'population2010', 'AGE135214', 'AGE295214', 'age65plus', 'White', 'Black', 'Hispanic', 'NonEnglish', 'Edu_highschool', 'Edu_batchelors', 'SEX255214', 'VET605213', 'Income', 'Poverty', '16Winner', '12Winner']]
votes.head(5)
```

To get a better idea which features will be best for a predictive model, we can do feature selection using an Extra Tree Classifier to identify which attributes contribute the most to predicting the 2016 election outcome. Using feature importance algorithms, we will be able to discover the best features to use in our classification model.

# Feature Selection

```python
#Feature Importance with Extra Tree Classifier
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier

#Create Feature/Label Dataframes
X = votesDF[['population2014', 'population2010', 'AGE135214', 'AGE295214', 'age65plus', 'White', 'Black', 'Hispanic', 'NonEnglish', 'Edu_highschool', 'Edu_batchelors', 'SEX255214', 'VET605213', 'Income', 'Poverty', '12Winner']]
Y = votesDF[['16Winner']]
X = np.array(X)
Y = np.array(Y) 

#Feature Extraction
model = ExtraTreesClassifier()
model.fit(X, Y)
print("Model Feature Importance:", model.feature_importances_)
type(model.feature_importances_)

import seaborn as sns

%matplotlib inline
plt.figure(figsize=(23,10))
plt.title('Feature Importance with Extra Tree Classifier')
plt.xlabel('Features')
plt.ylabel('Score')
ETCFeatures = ['population2014', 'population2010', 'AGE135214', 'AGE295214', 'age65plus', 'White', 'Black', 'Hispanic', 'NonEnglish', 'Edu_highschool', 'Edu_batchelors', 'SEX255214', 'VET605213', 'Income', 'Poverty', '12Winner']
sns.barplot(ETCFeatures, model.feature_importances_, palette='Blues_d')
```

We are given an importance score for each attribute. The larger score the more important the attribute. Based on the Extra Tree Classifier, the rankings for high scoring features as well as there score are as follows:

1: 2012 Election (.461)
2: Batchelors Education (.119)
3: % County Population White (.072)
4: % County Population Black (.059)
5: Veterans (.035)

We see the 2012 Election results are the best feature for predicting the 2016 election outcome by a large margin. In another notebook, we will look at county turnover where a county who voted Democrat switches to Republican and visa-versa, but for now we will be sure to use that feature in our prediction models.

We can also model feature selection using Recursive Feature Elimination to identify which attributes contribute the most to predicting the 2016 election outcome.

```python
#Feature Extraction with RFE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

#Feature Extraction using RFE with the logistic Regression Algorithm
model = LogisticRegression()
rfe = RFE(model, 1)
rfe = rfe.fit(X, Y)


print("RFE n_Features:", rfe.n_features_)
print("RFE support:", rfe.support_)
print("RFE ranking: ", rfe.ranking_)
```

The RFE model provides us with a ranking for the features. The top 5 rankings are as follows:

1 - 2012 Election outcome
2 - Hispanic population
3 - White population
4 - Black population
5 - Batchelors Education

The top 5 features has the top 4 using Extra Tree Classifier, but in a different ranking. Something interesting is that Veterans, the feature scoring fifth highest in the ETC, is ranked last using the RFE model. Given that it scored highly in one feature selection algorithm but dead last in the other, I'm going to omit using it as a feature in my Decision Tree Classifier.

# Decision Tree Classifier

Now that we know which features are the best to utilize in our ML models, I will use a decision tree classifier on those select features to predict the 2016 election outcomes. To get a sense of how well each feature performs, I will create different classifieres using each feature, and then will combine them to derive additional insights, and predict with higher accuracy.

```python
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree

labels = votes['16Winner']
features = votes[['Edu_batchelors']]
X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    labels, 
                                                    test_size=0.20, 
                                                    random_state=42)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(accuracy_score(y_test, clf.predict(X_test)))
```

Clearly the 2012 election winner for the county was the best predictor for the decision tree, as with that as the only training feature the model scores with a 92 percent accuracy. I am going to create models using multiple features to see how well the model develops.

```python
labels = votes['16Winner']
features = votes[['White', 'Black', 'Hispanic']]
X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    labels, 
                                                    test_size=0.20, 
                                                    random_state=42)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(accuracy_score(y_test, clf.predict(X_test)))
```

When combining the features looking at the % population are certain races, the model only predicts about 1 percent better than just using 1 race as a feature. This outcome is giving me a hunch that these variables may have a high co-correlation. These features here seem to classify in simailar ways, so combining them doesn't seem to provide too much additional insight.

```python
labels = votes['16Winner']
features = votes[['Edu_batchelors', 'White', 'Black', 'Hispanic']]
X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    labels, 
                                                    test_size=0.20, 
                                                    random_state=42)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(accuracy_score(y_test, clf.predict(X_test)))
```

After adding % population with a batchelors education, our model prediction increases by approximately 4 percent. 

```python
labels = votes['16Winner']
features = votes[['Edu_batchelors', 'White', 'Black', 'Hispanic', '12Winner']]
X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    labels, 
                                                    test_size=0.20, 
                                                    random_state=42)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(accuracy_score(y_test, clf.predict(X_test)))
```

Combining all our top performing features together, our decision tree classification model predicts with a 94.7 percent accuracy. While this is a highly accurate model, it is also important to note adding edu_batchelors, white, black, and hispanic as features only boosted the accuracy by about 2 percent. Using the 2012 winner alone, you can predict with high accuracy who counties are going to vote for. That being said, there is some voter turnover where counties voting one party vote the other the next election. While the percentage of counties doing this is rather small, it can make all the difference in a close election.
